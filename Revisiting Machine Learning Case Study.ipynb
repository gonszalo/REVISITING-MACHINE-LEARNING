```python
cols_to_remove = [col for col in data.columns if 'ADATE_' in col and col in cols_to_keep]
cols_to_remove
```

```python
[cols_to_keep.remove(col) for col in cols_to_remove]

cols_to_keep = list(set(cols_to_keep))
sorted(cols_to_keep)
```
